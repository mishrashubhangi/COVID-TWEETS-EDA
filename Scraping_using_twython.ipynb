```
users=[]
date=[]
texts=[]
fc=[]
pc=[]
for i in range(0,len(search_results['statuses'])):
  users.append(search_results['statuses'][i]['user']['name'])
  date.append(search_results['statuses'][i]['created_at'])
  texts.append(search_results['statuses'][i]['text'])
  fc.append(search_results['statuses'][i]['favorite_count'])
  pc.append(search_results['statuses'][i]['user']['location'])
```

```
data_new = {'User Name':  users,
            'Date of Tweet':date,
            'Text':texts,
            'Favourite Count':fc,
            'Location':pc
        } 
```

```
import pandas as pd
df_new = pd.DataFrame (data_new, columns = ['User Name','Date of Tweet','Text','Favourite Count','Location'])
```
